## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos_mat'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos_por'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_geo'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_ing'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_his'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos_fis'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos_qui'])
````

````python
# 1) Seu código aqui

# aqui crio uma nova coluna com o total de questoes para cada assunto

df_mat['qtd_questoes_mat'] = 24
df_por['qtd_questoes_por'] = 18
df_geo['qtd_questoes_geo'] = 8
df_ing['qtd_questoes_ing'] = 8
df_his['qtd_questoes_his'] = 8
df_fis['qtd_questoes_fis'] = 12
df_qui['qtd_questoes_qui'] = 12

# agora crio uma terceira coluna, mostrando a porcentagem de acertos de cada aluno, por assunto

df_mat['porc_acertos_mat'] = (df_mat['Qt_acertos_mat'] / df_mat['qtd_questoes_mat']*100).round(2)
df_por['porc_acertos_por'] = (df_por['Qt_acertos_por'] / df_por['qtd_questoes_por']*100).round(2)
df_geo['porc_acertos_geo'] = (df_geo['Qt_acertos_geo'] / df_geo['qtd_questoes_geo']*100).round(2)
df_ing['porc_acertos_ing'] = (df_ing['Qt_acertos_ing'] / df_ing['qtd_questoes_ing']*100).round(2)
df_his['porc_acertos_his'] = (df_his['Qt_acertos_his'] / df_his['qtd_questoes_his']*100).round(2)
df_fis['porc_acertos_fis'] = (df_fis['Qt_acertos_fis'] / df_fis['qtd_questoes_fis']*100).round(2)
df_qui['porc_acertos_qui'] = (df_qui['Qt_acertos_qui'] / df_qui['qtd_questoes_qui']*100).round(2)
````

````python
# 2) Seu código aqui

# o codigo abaixo soma todos os acertos por aluno em cada dataframe e mostra isso num novo DataFrame chamado 'total_acertos'

total_acertos = pd.DataFrame(df_mat['Qt_acertos_mat'] + df_por['Qt_acertos_por'] + df_geo['Qt_acertos_geo'] + df_ing['Qt_acertos_ing'] + df_his['Qt_acertos_his'] + df_fis['Qt_acertos_fis'] + df_qui['Qt_acertos_qui'], columns=['total_acertos'])
total_acertos.head()                                                                                                       
````

````python
# 3) Seu código aqui

acertos_porcentagem = pd.DataFrame({
    'acertos_porc': (total_acertos['total_acertos'] / 90).round(2)})
acertos_porcentagem.head()
````

````python
# 4) Seu código aqui
acima_45 = total_acertos[total_acertos['total_acertos'] > 45]
acima_45.count()
````

## BONUS



````python
# no codigo abaixo, combinarei todos os DataFrame criados acima em um unico DataFrame

df_vestibular_total = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui, total_acertos, acertos_porcentagem], axis=1)
df_vestibular_total.head()
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui

checagem_zero = ['Qt_acertos_mat', 'Qt_acertos_fis', 'Qt_acertos_qui']

df_zero = df_vestibular_total[(
    df_vestibular_total[checagem_zero] == 0).any(axis=1)]

df_zero[['Qt_acertos_mat', 'Qt_acertos_fis', 'Qt_acertos_qui']].head()
````

````python
# 2) Quantos são esses alunos?

total_zero = len(df_zero)
total_zero
````

````python
# 3) Qual a média desses alunos em história e geografia?

# aqui filtro as notas em historia e geografia de todos os alunos que tiraram zero em pelo menos uma das tres materias (fis, mat, qui)
# guardo essa media em duas variaveis 'media_his' e 'media_geo'
media_his = df_zero['Qt_acertos_his'].mean()
media_geo = df_zero['Qt_acertos_geo'].mean()

# resultados impressos
print(f"Média de Historia: {media_his:.2f}")
print(f"Média de Geografia: {media_geo:.2f}")
````

````python
# 4) Monte um DataFrame com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

df_alunos_classificados = df_vestibular_total[(
    df_vestibular_total[checagem_zero] != 0).all(axis=1)]

df_alunos_classificados.head()
````
